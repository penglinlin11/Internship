```python
# 1) Write a python program to display all the header tags from wikipedia.org and make data frame.

!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page

soup = BeautifulSoup(page.content)
soup

header_tag = []

for i in soup.find_all('span', class_="mw-headline"):
  header_tag.append(i.text)

header_tag

import pandas as pd

df = pd.DataFrame({'Header_tag':header_tag})
df
```

```python
# 2) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office)
# from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

page = requests.get("https://presidentofindia.nic.in/former-presidents")
page

soup = BeautifulSoup(page.content)
soup

data = {'name':[], 'Term_of_office':[]}

for i in soup.find_all('div', class_="desc-sec"):
  children = i.find_all(['h3','h5'])

  data['name'].append(children[0].text)  # Assuming the first child is the name
  data['Term_of_office'].append(children[1].text)

data


import pandas as pd

df = pd.DataFrame(data)
df
```

```python
# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
# i) Headline
# ii) Time
# iii) NewsLink

page = requests.get("https://www.cnbc.com/world/?region=world")
page

soup = BeautifulSoup(page.content)
soup

Headline = []

for i in soup.find_all('a', class_="LatestNews-headline"):
  Headline.append(i.text)

Headline

Time = []

for i in soup.find_all('time', class_="LatestNews-timestamp"):
  Time.append(i.text)

Time

NewsLink = []

for i in soup.find_all('a', href=True):
  NewsLink.append(i['href'])

NewsLink
```

```python
#6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
#i) Paper Title
#ii) Authors
#iii) Published Date
#iv) Paper URL


!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

page = requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
page

soup = BeautifulSoup(page.content)
soup

Paper_Title = []

for i in soup.find_all('h2', class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
  Paper_Title.append(i.text)

Paper_Title

Authors = []

for i in soup.find_all('span', class_="sc-1w3fpd7-0 dnCnAO"):
  Authors.append(i.text)

Authors

Published_Date = []

for i in soup.find_all('span', class_="sc-1thf9ly-2 dvggWt"):
  Published_Date.append(i.text)

Published_Date

Paper_URL = []

for i in soup.find_all('a', class_="sc-5smygv-0 fIXTHm"):
  Paper_URL.append(i.text)

Paper_URL

import pandas as pd

df = pd.DataFrame({'Paper_Title':Paper_Title, 'Authors':Authors,'Published_Date':Published_Date,'Paper_URL':Paper_URL})
df
```

```python
#7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-
#i) Restaurant_name
#ii) Cuisine
#iii) Location
#iv) Ratings
#v) Image_URL


page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page

soup = BeautifulSoup(page.content)
soup

Restaurant_name = []

for i in soup.find_all('a', class_="restnt-name ellipsis"):
  Restaurant_name.append(i.text)

Restaurant_name

Cuisine = []

for i in soup.find_all('span', class_="double-line-ellipsis"):
  Cuisine.append(i.text)

Cuisine

Location = []

for i in soup.find_all('div', class_="restnt-loc ellipsis"):
  Location.append(i.text)

Location

Ratings = []

for i in soup.find_all('div', class_="restnt-rating rating-4"):
  Ratings.append(i.text)

Ratings

Image_URL = []

for i in soup.find_all('img', class_="no-img"):
  Image_URL.append(i['data-src'])

Image_URL

print(len(Restaurant_name),len(Cuisine),len(Location),len(Ratings),len(Image_URL))

import pandas as pd

df = pd.DataFrame({'Restaurant_name':Restaurant_name, 'Cuisine':Cuisine, 'Location':Location,'Ratings':Ratings,'Image_URL':Image_URL})
df

```
